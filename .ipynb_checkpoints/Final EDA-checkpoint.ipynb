```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('cleaneddata.csv')
```

```python
df['churn'].value_counts()
```

## 1. Is  the company doing really well in this market 

```python
print('Customers to be churn and churn out(0:No Churn, 1:Churn)')
print(df['churn'].value_counts())
sns.countplot(x='churn', data=df, palette='muted')
plt.xticks(ticks = [1,0], labels=['churn' ,'non_churn'])
plt.show()

c= df.churn.sum()
nc= len(df)-c

cperc= round(c/len(df)*100,2)
ncperc= round(nc/len(df)*100,2)

print("Customers to be no churn out are:- ",ncperc,'%')
print("Customers to be churn out are:- ",cperc,'%')
```

```python
#Churn Distribution
ax = (df['churn'].value_counts()*100.0 /len(df)).plot.pie(autopct='%.1f%%', labels = ['No', 'Yes'],figsize =(5,5), fontsize = 12 )                                                                           

ax.set_ylabel('Churn',fontsize = 12)
ax.set_title('Churn Distribution', fontsize = 12)
```

- The above graph suggests that yes they are doing really well when it comes to churning of customers 
- only 08.14 % customers are churning out and 91.86 % are interested in continuing the service 

## 2. Number of Customers calling on the same Network (ONNET)

```python
def fun1_eda(telecomCols):
    plt.rcParams["figure.figsize"] = [15, 6]
    fig = plt.figure()
    cols = {c: telecomCols + "_" + str(c) for c in [6,7,8]}
    print(cols)
    f,axes = plt.subplots(nrows=1, ncols=3)
    sns.boxplot(data=df, y=telecomCols + "_6", x="churn",
                hue="churn", linewidth=2.5, showfliers=False, ax=axes[0])
    sns.boxplot(data=df, y=telecomCols + "_7",  x="churn",
                hue="churn", linewidth=2.5, showfliers=False, ax=axes[1])
    sns.boxplot(data=df, y=telecomCols + "_8",  x="churn",
                hue="churn", linewidth=2.5, showfliers=False, ax=axes[2]);
```

```python
fun1_eda('onnet_mou')
```

## 3. Number of Customers calling on the different Network (OFFNET)

```python
fun1_eda('offnet_mou')
```

## 7. How recharge amount can help you to predict churn customers?

```python
def fun3_eda(telecomCols):
    plt.rcParams["figure.figsize"] = [15, 6]
    fig = plt.figure()
    cols = {c: telecomCols + "_" + str(c) for c in [6,7,8]}
    print(cols)
    f,axes = plt.subplots(nrows=1, ncols=3)
    sns.boxplot(data=df, y=telecomCols + "_6", x="churn",
                hue="churn", linewidth=2.5, showfliers=False, ax=axes[0])
    sns.boxplot(data=df, y=telecomCols + "_7",  x="churn",
                hue="churn", linewidth=2.5, showfliers=False, ax=axes[1])
    sns.boxplot(data=df, y=telecomCols + "_8",  x="churn",
                hue="churn", linewidth=2.5, showfliers=False, ax=axes[2]);
```

```python
# let's analyze total recharge amount:
fun3_eda('total_rech_amt')
```

```python
# let's analyze Average recharge amount:
fun3_eda('av_rech_amt_data')
```

* Above three graphs clearly shows that
 - For total_rech_amt Churn rate is decreasing in nature.  
 - For av_rech_amt_data Churn rate is drastically decreasing.
